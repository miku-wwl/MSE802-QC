# Statevector, Probability Amplitudes, Ket and Bra Notation

This notebook introduces the core mathematical notation used in quantum computing:

- Statevectors
- Probability amplitudes
- Ket notation
- Bra notation
- Inner products
- Outer products
- Tensor products
- One-qubit and two-qubit quantum states

The notebook is designed for students learning the foundations of quantum computing.

```python
import numpy as np
```

## 1. What is a Quantum State?

In quantum computing, the state of a qubit is written using **ket notation**:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

Here, $\alpha$ and $\beta$ are called **probability amplitudes**.

They can be real or complex numbers.

The computational basis states are:

$$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
|1\rangle =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

```python
# Define basis states |0> and |1>

# Define basis states |0> and |1>
ket_0 = np.array([[1],
                  [0]])

ket_1 = np.array([[0],
                  [1]])

print("|0> =")
print(ket_0)

print("\n|1> =")
print(ket_1)
```

## 2. Statevector Representation

A general single-qubit state is:

$$
|\psi\rangle =
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
$$

This column vector is called a **statevector**.

For a valid quantum state, the probabilities must add to 1:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

This is called the **normalisation condition**.

```python
# Example state:
# |ψ> = 1/sqrt(2)|0> + 1/sqrt(2)|1>

alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)

psi = np.array([[alpha],
                [beta]])

print("Statevector |ψ> =")
print(psi)

# Check normalisation
normalisation = abs(alpha)**2 + abs(beta)**2
print("\n Normalisation =", normalisation)
```

## 3. Probability Amplitudes

In the state:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

The probability of measuring $|0\rangle$ is:

$$
P(0) = |\alpha|^2
$$

The probability of measuring $|1\rangle$ is:

$$
P(1) = |\beta|^2
$$

The amplitude itself is not the probability. The probability is the squared magnitude of the amplitude.

```python
prob_0 = abs(alpha)**2
prob_1 = abs(beta)**2

print("Probability of measuring |0> =", prob_0)
print("Probability of measuring |1> =", prob_1)
print("Total probability =", prob_0 + prob_1)
```

## 4. Measurement Simulation

A quantum measurement gives one classical outcome.

For a single qubit, the possible outcomes are:

$$
0
$$

or

$$
1
$$

The probabilities depend on the squared magnitudes of the amplitudes.

```python
# Simulate measurement many times

outcomes = [0, 1]
probabilities = [prob_0, prob_1]

shots = 1000
measurements = np.random.choice(outcomes, size=shots, p=probabilities)

count_0 = np.sum(measurements == 0)
count_1 = np.sum(measurements == 1)

print("Measurement results after", shots, "shots:")
print("0 measured:", count_0)
print("1 measured:", count_1)
```

## 5. Bra Notation

A ket is a column vector:

$$
|\psi\rangle =
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
$$

The corresponding bra is the conjugate transpose:

$$
\langle \psi| =
\begin{bmatrix}
\alpha^* & \beta^*
\end{bmatrix}
$$

Here, $^*$ means complex conjugate.

If a number is complex, for example:

$$
i
$$

then its complex conjugate is:

$$
-i
$$

```python
# Define a state with a complex amplitude

alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)

ket_psi = np.array([[alpha],
                    [beta]])

bra_psi = ket_psi.conj().T

print("Ket |ψ> =")
print(ket_psi)

print("Bra <ψ| =")
print(bra_psi)
```

## 6. Inner Product

The inner product is written as:

$$
\langle \psi | \psi \rangle
$$

For a valid quantum state:

$$
\langle \psi | \psi \rangle = 1
$$

This confirms that the state is normalised.

```python
inner_product = bra_psi @ ket_psi

print("<ψ|ψ> =")
print(inner_product)
```

## 7. Outer Product

The outer product is written as:

$$
|\psi\rangle \langle \psi|
$$

This creates a matrix representation of the state. For a pure quantum state, this matrix is also called a **density matrix**.

```python
outer_product = ket_psi @ bra_psi

print("|ψ><ψ| =")
print(outer_product)
```

## 8. Two-Qubit Statevector

For two qubits, the computational basis states are:

$$
|00\rangle, |01\rangle, |10\rangle, |11\rangle
$$

A general two-qubit state is:

$$
|\psi\rangle =
\alpha |00\rangle +
\beta |01\rangle +
\gamma |10\rangle +
\delta |11\rangle
$$

The statevector is:

$$
|\psi\rangle =
\begin{bmatrix}
\alpha \\
\beta \\
\gamma \\
\delta
\end{bmatrix}
$$

The normalisation condition is:

$$
|\alpha|^2 + |\beta|^2 + |\gamma|^2 + |\delta|^2 = 1
$$

```python
# Example Bell state:
# |Φ+> = 1/sqrt(2)|00> + 1/sqrt(2)|11>

alpha = 1 / np.sqrt(2)
beta = 0
gamma = 0
delta = 1 / np.sqrt(2)

bell_state = np.array([[alpha],
                       [beta],
                       [gamma],
                       [delta]])

print("Bell state |Φ+> =")
print(bell_state)

probabilities = np.abs(bell_state.flatten())**2
basis_states = ["00", "01", "10", "11"]

print("Measurement probabilities:")
for state, prob in zip(basis_states, probabilities):
    print(f"P({state}) = {prob}")
```

## 9. Tensor Product

To combine quantum states, we use the tensor product.

For example:

$$
|0\rangle \otimes |0\rangle = |00\rangle
$$

$$
|0\rangle \otimes |1\rangle = |01\rangle
$$

$$
|1\rangle \otimes |0\rangle = |10\rangle
$$

$$
|1\rangle \otimes |1\rangle = |11\rangle
$$

The tensor product allows us to build multi-qubit systems from single-qubit states.

```python
ket_00 = np.kron(ket_0, ket_0)
ket_01 = np.kron(ket_0, ket_1)
ket_10 = np.kron(ket_1, ket_0)
ket_11 = np.kron(ket_1, ket_1)

print("|00> =")
print(ket_00)

print("|01> =")
print(ket_01)

print("|10> =")
print(ket_10)

print("|11> =")
print(ket_11)
```

## 10. Visualising Measurement Probabilities

We can plot the measurement probabilities for a quantum state.

For the Bell state:

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle
$$

Only $|00\rangle$ and $|11\rangle$ have non-zero probabilities.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.bar(basis_states, probabilities)
plt.xlabel("Basis state")
plt.ylabel("Probability")
plt.title("Measurement Probabilities for Bell State")
plt.ylim(0, 1)
plt.show()
```

## 11. Key Summary

A quantum state is represented as a statevector.

For one qubit:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

For two qubits:

$$
|\psi\rangle =
\alpha |00\rangle +
\beta |01\rangle +
\gamma |10\rangle +
\delta |11\rangle
$$

Probability is found by squaring the magnitude of each amplitude:

$$
P(i) = |\text{amplitude}_i|^2
$$

A ket is a column vector:

$$
|\psi\rangle
$$

A bra is the conjugate transpose:

$$
\langle \psi|
$$

The inner product checks normalisation:

$$
\langle \psi | \psi \rangle = 1
$$
